<h3>Part 1: Load the data</h3>

```python
import numpy as np
import matplotlib.pyplot as plt

with open("data_example.txt") as file:
    container = [each.rstrip() for each in file]
    file.close()

population = []
profit = []

for each in container:
    temp = each.split(",")
    population.append(float(temp[0]))
    profit.append(float(temp[1]))


#convert data into matrix
ppl_array = np.array(population)
profit_array = np.array(profit)

#plot the data
%matplotlib inline
plt.rcParams['figure.figsize'] = [8, 6]
plt.scatter(population, profit, color='red', label='Real data', alpha=0.75)
plt.grid(color='#333', ls='--', lw=0.5)
plt.legend(loc='upper left')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

<h3>Part 2: Fit Model to Data</h3>

```python
import gradient_descent as gds

model = gds.gradient_des(ppl_array, profit_array)
model.set_params(0.01, [0,0])
num_iter, j_iter = model.fit_model(1500)
theta = model.get_params()
cost, y_pred = model.pred_cost()
```

<h3>Part 3: Plot Model with Data</h3>

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [8, 6]
plt.scatter(population, profit, color='red', label='Real data', alpha=0.75)
plt.plot(population, y_pred, color='black', label='Regression fit')
plt.grid(color='#333', ls='--', lw=0.5)
plt.legend(loc='upper left')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()

%matplotlib inline
plt.rcParams['figure.figsize'] = [8, 6]
plt.plot(num_iter, j_iter, color='blue', label='Cost Function', lw=3)
plt.grid(color='#333', ls='--', lw=0.5)
plt.legend(loc='upper left')
plt.xlabel('Number of Iteration')
plt.ylabel('Cost (J)')
plt.show()
```

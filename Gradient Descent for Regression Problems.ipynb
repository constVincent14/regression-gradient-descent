<h3>Example How To Use The Code</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Dataset
with open('from_XOR_clockwise.txt', 'r') as file:
    container = [each.rstrip() for each in file]
    file.close()

temp = []
for index, value in enumerate(container):
    if index==0:
        continue
    temp.append(value.split("\t"))

rotation = []
pulse = []
for each in temp:
    rotation.append(float(each[0]))
    pulse.append(float(each[-1])) 

rotation = np.array(rotation)
pulse = np.array(pulse)
```

```python
import gradient_descent as gds

grad_model = gds.gradient_des(rotation, pulse)
grad_model.set_params(learn_rate=0.0001, thetas=[-4, -407])
grad_model.fit_model(15)
theta = grad_model.show_params()

%matplotlib inline
plt.scatter(rotation, pulse, color='red', label='DATA')
plt.plot(rotation, [theta[0]+(theta[1]*value) for value in rotation],color='black', label='LEAST SQUARE')
plt.grid(c='#333', lw=0.5, ls='--')
plt.xlabel("ROTATION")
plt.ylabel("PULSE")
plt.legend()
plt.show()
```
